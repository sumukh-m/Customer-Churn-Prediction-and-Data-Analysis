# Customer Churn Prediction and Data Analysis

## Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Load Dataset

```python
# Load the CSV file into a pandas DataFrame
file_path = 'Bank_Churn.csv'
df = pd.read_csv(file_path)
```

## Exploratory Data Analysis

```python
# Display basic information about the dataset
print("Dataset Information:")
print(df.info())
```

```python
# Display summary statistics of numerical columns
print("\nSummary Statistics:")
print(df.describe())
```

```python
# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Explore categorical variables
categorical_columns = ['GeographyID', 'GenderID', 'NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Exited']
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df, x=column, hue='Exited')
    plt.title(f"{column} Distribution")
    plt.show()
```

```python
# Explore correlation between numerical variables
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Explore age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Explore tenure distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Tenure', bins=10, kde=True)
plt.title("Tenure Distribution")
plt.show()
```

```python
# Explore balance distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Balance', bins=20, kde=True)
plt.title("Balance Distribution")
plt.show()
```

```python
# Explore salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='EstimatedSalary', bins=20, kde=True)
plt.title("Estimated Salary Distribution")
plt.show()
```
